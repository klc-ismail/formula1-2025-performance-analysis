# Part 2 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
laps = pd.read_csv("master_laps.csv")
stints = pd.read_csv("master_stints.csv")

laps = laps.sort_values(["session_key", "driver_number", "lap"])
stints = stints.sort_values(["session_key", "driver_number", "stint"])
```

```python
laps_clean = laps[
    (laps["is_clean_lap"] == True) &
    (laps["lap_time"].notna())
].copy()
```

```python
stints["lap_start"] = stints["lap_start"].astype("Int64")
stints["lap_end"]   = stints["lap_end"].astype("Int64")
```

```python
def degradation_slope(laps_in_stint: pd.DataFrame) -> float:
    df = laps_in_stint.dropna(subset=["lap_time", "lap_in_stint"]).copy()
    if len(df) < 5:  # for slope reliability in very short stints
        return np.nan

    x = df["lap_in_stint"].astype(float).to_numpy()
    y = df["lap_time"].astype(float).to_numpy()

    a, b = np.polyfit(x, y, 1)
    return float(a)
```

```python
rows = []

laps_perf = laps_clean[[
    "session_key", "track", "team", "driver", "driver_number",
    "lap", "lap_time"
]].copy()

for _, s in stints.iterrows():
    sk = s["session_key"]
    dn = s["driver_number"]
    lap_start = s["lap_start"]
    lap_end = s["lap_end"]

    if pd.isna(lap_start) or pd.isna(lap_end):
        continue

    l = laps_perf[
        (laps_perf["session_key"] == sk) &
        (laps_perf["driver_number"] == dn) &
        (laps_perf["lap"] >= int(lap_start)) &
        (laps_perf["lap"] <= int(lap_end))
    ].copy()

    if l.empty:
        continue

    l = l.sort_values("lap")
    l["lap_in_stint"] = np.arange(1, len(l) + 1)

    rows.append({
        "session_key": sk,
        "track": l["track"].iloc[0],
        "team": l["team"].iloc[0],
        "driver": l["driver"].iloc[0],
        "driver_number": int(dn),
        "stint": int(s["stint"]),
        "compound": s.get("compound", np.nan),

        "lap_start": int(lap_start),
        "lap_end": int(lap_end),
        "stint_laps_clean": int(len(l)),

        "median_clean_lap_time": float(l["lap_time"].median()),
        "best_clean_lap_time": float(l["lap_time"].min()),
        "std_clean_lap_time": float(l["lap_time"].std(ddof=1)) if len(l) > 1 else np.nan,

        "degradation_slope": degradation_slope(l)
    })

df_stint = pd.DataFrame(rows).sort_values(
    ["session_key", "team", "driver_number", "stint"]
).reset_index(drop=True)
```

```python
df_stint_q = df_stint[df_stint["stint_laps_clean"] >= 5].copy()
```

```python
df_pace = (
    df_stint_q
    .groupby(["session_key", "track", "team", "driver", "driver_number"], as_index=False)
    .agg(
        median_race_pace=("median_clean_lap_time", "median"),
        best_race_pace=("best_clean_lap_time", "min"),
        consistency=("std_clean_lap_time", "mean"),
        degradation=("degradation_slope", "mean"),
        total_clean_laps=("stint_laps_clean", "sum")
    )
)

# DNF/az veri filtre
df_pace = df_pace[df_pace["total_clean_laps"] >= 10].copy()

df_pace = df_pace.sort_values(["session_key", "team", "median_race_pace"]).reset_index(drop=True)
```

```python
rows = []

for (sk, team), g in df_pace.groupby(["session_key", "team"]):
    if len(g) != 2:
        continue

    g = g.sort_values("median_race_pace")
    faster = g.iloc[0]
    slower = g.iloc[1]

    rows.append({
        "session_key": sk,
        "track": faster["track"],
        "team": team,
        "pace_diff_sec": float(slower["median_race_pace"] - faster["median_race_pace"])
    })

df_teammate = pd.DataFrame(rows)
```

```python
df_teammate.head()
```

```python
TEAM_COLORS = {
    "RedBull": "#1E41FF",
    "Ferrari": "#DC0000",
    "McLaren": "#FF8700",
    "Mercedes": "#00D2BE",
    "AstonMartin": "#006F62"
}
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df_teammate,
    x="team",
    y="pace_diff_sec",
    palette=TEAM_COLORS
)
plt.title("Teammate Pace Difference Distribution (Average Clean Lap) by Team")
plt.ylabel("Pace Difference (maximum) (Largest = greater intra-team difference)")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
df_strategy_stint = df_stint_q.copy()

df_strategy_stint["stint_length"] = (
    df_strategy_stint["lap_end"] - df_strategy_stint["lap_start"] + 1
)

# Basic quality filter (shorter stints out)
df_strategy_stint = df_strategy_stint[
    df_strategy_stint["stint_length"] >= 5
].copy()
```

```python
df_strategy_stint[[
    "team", "driver", "stint", "stint_length", "median_clean_lap_time"
]].head()
```

## graph 1 

```python
g = sns.relplot(
    data=df_strategy_stint,
    x="stint_length",
    y="median_clean_lap_time",
    col="team",
    col_wrap=3,
    kind="scatter",
    hue="team",
    palette=TEAM_COLORS,
    alpha=0.6,
    height=4,
    facet_kws={"sharex": True, "sharey": True}
)

# LOWESS trend her panel iin
for ax, team in zip(g.axes.flatten(), g.col_names):
    team_df = df_strategy_stint[df_strategy_stint["team"] == team]
    sns.regplot(
        data=team_df,
        x="stint_length",
        y="median_clean_lap_time",
        scatter=False,
        lowess=True,
        ax=ax,
        color="black",
        line_kws={"linewidth": 2}
    )

g.set_axis_labels(
    "Stint Length (laps)",
    "Average Clean Lap Pace (sec)"
)

g.fig.suptitle(
    "Relationship Between Stint Length and Race Pace \n (A Team-Based Strategic Approach)",
    y=1.03
)

g._legend.remove()

plt.show()
```

## Graph 2 

```python
pit = pd.read_csv("master_pit.csv")

pit = pit.sort_values(["session_key", "driver_number", "lap"])
laps_clean = laps_clean.sort_values(["session_key", "driver_number", "lap"])
```

```python
race_laps_df = (
    laps_clean
    .groupby("session_key", as_index=False)
    .agg(race_laps=("lap", "max"))
)
```

```python
def compute_pit_window(
    laps_df,
    pit_lap,
    session_key,
    driver_number,
    window=3
):
    before = laps_df[
        (laps_df["session_key"] == session_key) &
        (laps_df["driver_number"] == driver_number) &
        (laps_df["lap"] < pit_lap) &
        (laps_df["lap"] >= pit_lap - window)
    ]

    after = laps_df[
        (laps_df["session_key"] == session_key) &
        (laps_df["driver_number"] == driver_number) &
        (laps_df["lap"] > pit_lap) &
        (laps_df["lap"] <= pit_lap + window)
    ]

    if len(before) < 2 or len(after) < 2:
        return None

    return {
        "pace_before": before["lap_time"].mean(),
        "pace_after": after["lap_time"].mean(),
        "delta_pace": after["lap_time"].mean() - before["lap_time"].mean()
    }
```

```python
rows = []

for _, p in pit.iterrows():
    res = compute_pit_window(
        laps_df=laps_clean,
        pit_lap=p["lap"],
        session_key=p["session_key"],
        driver_number=p["driver_number"],
        window=3
    )

    if res is None:
        continue

    rows.append({
        "session_key": p["session_key"],
        "track": p["track"],
        "team": p["team"],
        "driver": p["driver"],
        "driver_number": p["driver_number"],
        "pit_lap": p["lap"],
        "pit_duration": p["pit_duration"],
        "pace_before": res["pace_before"],
        "pace_after": res["pace_after"],
        "delta_pace": res["delta_pace"]
    })

pit_window_df = pd.DataFrame(rows)
```

```python
pit_window_df = pit_window_df.merge(
    race_laps_df,
    on="session_key",
    how="left"
)

pit_window_df["pit_phase"] = (
    pit_window_df["pit_lap"] / pit_window_df["race_laps"]
)
```

```python
pit_window_df = pit_window_df[
    pit_window_df["delta_pace"].between(-10, 10)
].copy()
```

```python
pit_window_df.head()
```

```python
g = sns.relplot(
    data=pit_window_df,
    x="pit_phase",
    y="delta_pace",
    col="team",
    col_wrap=3,
    kind="scatter",
    hue="team",
    palette=TEAM_COLORS,
    alpha=0.6,
    height=4,
    facet_kws={"sharex": True, "sharey": True}
)

# LOWESS trend  for each team
for ax, team in zip(g.axes.flatten(), g.col_names):
    team_df = pit_window_df[pit_window_df["team"] == team]
    sns.regplot(
        data=team_df,
        x="pit_phase",
        y="delta_pace",
        scatter=False,
        lowess=True,
        ax=ax,
        color="black",
        line_kws={"linewidth": 2}
    )

# Reference line (0 = no tempo change)
for ax in g.axes.flatten():
    ax.axhline(0, color="gray", linestyle="--", linewidth=1)

g.set_axis_labels(
    "Pit Timing (0 = Early, 1 = Late)",
    "Tempo Change (sec/lap)\n(Negative = Faster)"
)

g.fig.suptitle(
    "Pit Stop Timing and Pace Gain \n (Strategic Impact on Teams)",
    y=1.03
)

g._legend.remove()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from matplotlib.patches import Patch
```

```python
laps = pd.read_csv("master_laps.csv")
pit = pd.read_csv("master_pit.csv")
summary = pd.read_csv("master_summary.csv")
```

```python
laps = laps.sort_values(["session_key", "driver_number", "lap"])
pit = pit.sort_values(["session_key", "driver_number", "lap"])

laps_clean = laps[
    (laps["is_clean_lap"] == True) &
    (laps["lap_time"].notna())
].copy()
```

## Team-based bit rate and stability.

```python
team_pit_stats = (
    pit
    .groupby("team")
    .agg(
        pit_count=("pit_duration", "count"),
        mean_pit_duration=("pit_duration", "mean"),
        median_pit_duration=("pit_duration", "median"),
        std_pit_duration=("pit_duration", "std"),
        q1=("pit_duration", lambda x: x.quantile(0.25)),
        q3=("pit_duration", lambda x: x.quantile(0.75)),
    )
    .reset_index()
)

team_pit_stats["iqr_pit_duration"] = (
    team_pit_stats["q3"] - team_pit_stats["q1"]
)

team_pit_stats = team_pit_stats.sort_values("median_pit_duration")
```

```python
TEAM_COLORS = {
    "RedBull": "#3671C6",      # Red Bull Blue
    "Ferrari": "#E80020",      # Ferrari Red
    "McLaren": "#FF8000",      # McLaren Orange
    "Mercedes": "#27F4D2",     # Mercedes Teal
    "AstonMartin": "#229971"   # Aston Martin Green
}
```

```python
plt.figure(figsize=(13, 6))

# Violin plot
sns.violinplot(
    data=pit,
    x="team",
    y="pit_duration",
    order=team_pit_stats["team"],
    palette=TEAM_COLORS,
    inner=None,
    alpha=0.6
)

# Boxplot (on violin plot)
sns.boxplot(
    data=pit,
    x="team",
    y="pit_duration",
    order=team_pit_stats["team"],
    width=0.25,
    showcaps=True,
    boxprops={"facecolor": "white", "zorder": 3},
    showfliers=True
)

plt.xticks(rotation=45)
plt.title("Pit Stop Duration Distribution by Team")
plt.ylabel("Pit Time (sec)")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

#  Pit Stop Strategy: 1-Stop / 2-Stop

```python
strategy_df = (
    summary
    .dropna(subset=["pit_count"])
    .assign(
        strategy=lambda df: np.where(df["pit_count"] == 1, "1-stop", "2-stop+")
    )
)

team_strategy_stats = (
    strategy_df
    .groupby(["team", "strategy"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

team_strategy_stats["total"] = (
    team_strategy_stats["1-stop"] + team_strategy_stats["2-stop+"]
)

team_strategy_stats["one_stop_rate"] = (
    team_strategy_stats["1-stop"] / team_strategy_stats["total"]
)

team_strategy_stats["two_stop_rate"] = (
    team_strategy_stats["2-stop+"] / team_strategy_stats["total"]
)
```

## Stacked Bar (Strategy Deployment)

```python
def adjust_color(color, factor=1.3):
    r, g, b = mcolors.to_rgb(color)
    return (min(r*factor,1), min(g*factor,1), min(b*factor,1))

legend_elements = []
teams = team_strategy_stats["team"]

for team in teams:
    base_color = TEAM_COLORS[team]
    dark_color = adjust_color(base_color, 0.75)

    legend_elements.append(
        Patch(facecolor=base_color, label=f"{team}  1-stop")
    )
    legend_elements.append(
        Patch(facecolor=dark_color, label=f"{team}  2-stop+")
    )
```

```python
plt.figure(figsize=(10, 6))

teams = team_strategy_stats["team"]
x = range(len(teams))

one_stop = team_strategy_stats["one_stop_rate"]
two_stop = team_strategy_stats["two_stop_rate"]

# Takm renkleri
base_colors = [TEAM_COLORS[t] for t in teams]
dark_colors = [adjust_color(c, 0.75) for c in base_colors]

# 1-stop (alt)
plt.bar(
    x,
    one_stop,
    color=base_colors
)

# 2-stop+ (st)
plt.bar(
    x,
    two_stop,
    bottom=one_stop,
    color=dark_colors
)

plt.xticks(x, teams, rotation=45)
plt.ylabel("Rate")
plt.title("Team-Based Pit Stop Strategy Distribution")

#  Custom Legend
legend_elements = []
for team in teams:
    base_color = TEAM_COLORS[team]
    dark_color = adjust_color(base_color, 0.75)

    legend_elements.append(
        Patch(facecolor=base_color, label=f"{team}  1-stop")
    )
    legend_elements.append(
        Patch(facecolor=dark_color, label=f"{team}  2-stop+")
    )

plt.legend(
    handles=legend_elements,
    bbox_to_anchor=(1.02, 1),
    loc="upper left",
    title="Team & Strategy"
)

plt.tight_layout()
plt.show()
```

# Do pit stops really pick up the pace?

```python
def get_pre_post_pace(laps_df, session_key, driver_number, pit_lap, window=3):
    driver_laps = laps_df[
        (laps_df["session_key"] == session_key) &
        (laps_df["driver_number"] == driver_number)
    ]

    pre = driver_laps[driver_laps["lap"] < pit_lap].tail(window)["lap_time"]
    post = driver_laps[driver_laps["lap"] > pit_lap].head(window)["lap_time"]

    if len(pre) < window or len(post) < window:
        return np.nan, np.nan

    return pre.mean(), post.mean()
```

```python
pit_window_rows = []

for _, row in pit.iterrows():
    pre, post = get_pre_post_pace(
        laps_clean,
        row["session_key"],
        row["driver_number"],
        row["lap"]
    )

    if np.isnan(pre) or np.isnan(post):
        continue

    pit_window_rows.append({
        "team": row["team"],
        "driver": row["driver"],
        "pit_lap": row["lap"],
        "pit_duration": row["pit_duration"],
        "pre_pit_pace": pre,
        "post_pit_pace": post,
        "delta_pace": post - pre
    })

pit_window_df = pd.DataFrame(pit_window_rows)
```

```python
team_pit_stats = (
    pit
    .groupby("team")
    .agg(
        pit_count=("pit_duration", "count"),
        mean_pit_duration=("pit_duration", "mean"),
        median_pit_duration=("pit_duration", "median"),
        std_pit_duration=("pit_duration", "std"),
        q1=("pit_duration", lambda x: x.quantile(0.25)),
        q3=("pit_duration", lambda x: x.quantile(0.75)),
    )
    .reset_index()
)

team_pit_stats["iqr_pit_duration"] = (
    team_pit_stats["q3"] - team_pit_stats["q1"]
)

team_pit_stats = team_pit_stats.sort_values("median_pit_duration")
```

```python
sns.lmplot(
    data=pit_window_df,
    x="pit_duration",
    y="delta_pace",
    col="team",
    col_wrap=3,
    height=4,
    hue="team",
    palette=TEAM_COLORS,
    scatter_kws={"alpha": 0.6},
    ci=None,
    legend=False
)

plt.show()
```

# Team-Based Post-Pit Pace Gain

```python
team_delta_stats = (
    pit_window_df
    .groupby("team")
    .agg(
        median_delta_pace=("delta_pace", "median"),
        mean_delta_pace=("delta_pace", "mean")
    )
    .sort_values("median_delta_pace")
    .reset_index()
)
```

## Bar Chart (Median Tempo Gain by Team)

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=team_delta_stats,
    x="median_delta_pace",
    y="team",
    palette=TEAM_COLORS
)

plt.axvline(0, color="black", linestyle="--", linewidth=1)
plt.title("Team-Based Average Speed Change After Pit Stop.")
plt.xlabel("Average Speed Change (sec/lap)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
